```python
# Author: Mahesh Nair
# Date: 3/21/2017

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.svm import LinearSVC, SVC
from sklearn import metrics
from sklearn.neural_network import MLPClassifier

%matplotlib inline

test_pix = pd.read_csv("test.csv")
train_pix = pd.read_csv("train.csv")
```

```python
# Splitting data in to X and y for attributes and its corresponding label output
y_train=train_pix['label']
X_train=train_pix
X_test=test_pix

# deleting the 'label' column from X as it is maintained in y
del X_train['label']

X_train.head()
```

```python
# clf=SVC(kernel='rbf', gamma=2, C=1)
# clf.fit(X,y)

#y_pred=clf.predict(X)

#print(y_pred)

clf = LinearSVC()
clf.fit(X_train,y_train)
```

```python
y_pred=clf.predict(X_train)

accuracy=metrics.accuracy_score(y_train,y_pred)
print(accuracy)
```

```python
mlp = MLPClassifier(hidden_layer_sizes=(6,6,6))
mlp.fit(X_train,y_train)
predictions = mlp.predict(X_train)
print(metrics.accuracy_score(y_train,predictions))
```
